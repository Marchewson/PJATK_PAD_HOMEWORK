Wgraj zawartość pliku orders.csv do ramki danych.
Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd
import numpy as np
df_orders = pd.read_csv('orders.csv')
df_orders.head(10)
```

```python
df_orders.info()
```

```python
df_orders.describe()
```

```python
#1.Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_orders
#df_orders.info()
```

```python
#2.Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
print('Unikalnych wartości jest: ', df_orders['tshirt_category'].nunique())
print('Unikalne wartości to: ', df_orders['tshirt_category'].unique())
```

```python
#3.Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

df_orders['tshirt_category'] = df_orders['tshirt_category'].apply(lambda x: 'White ' + x.split('Wh ', 1)[1] if x.startswith('Wh ') == True \
                                                                  else ('Black ' + x.split('Bl ', 1)[1] if x.startswith('Bl ') == True else x))

df_orders['tshirt_category'] = df_orders['tshirt_category'].apply(lambda x: x.split('Tshirt')[0] + 'T-Shirt' + x.split('Tshirt')[1] if 'Tshirt' in x else x)

print('Unikalne wartości to: ', df_orders['tshirt_category'].unique())
```

```python
#4.Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
df_orders['tshirt_type'] =  df_orders['tshirt_category'].apply(lambda x: x.split()[1] if x.count(' ') > 1 else x)
df_orders['tshirt_size'] =  df_orders['tshirt_category'].apply(lambda x: x.split()[2] if x.count(' ') > 1 else np.nan)
df_orders['tshirt_colour'] =  df_orders['tshirt_category'].apply(lambda x: x.split()[0] if x.count(' ') > 1 else np.nan)
df_orders
```

```python
#5.Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.
df_orders.loc[((df_orders['order_date'] >= '2014-12-31') & (df_orders['order_date'] <= '2016-08-02'))]
```

Wgraj plik customers.csv do ramki danych. 

Następnie:
1. Wypisz kolumny;
2. Porównaj rozmiar ramki z rozmiarem ramki w zadaniu 1;
3. Zmień nazwę kolumny customerID  na customer_id upewniając się, że dane zmiana zostanie zapisana w pamięci dzięki jednemu z argumentów metody do zmiany nazwy kolumny;
4. Połącz ramkę orders z ramką customers po kolumnie customer_id. Z której metody skorzystasz? Zostaw odpowiedź w komentarzu do zadania w kodzie (np. blok komentarzem ‘’’ ‘’’).


```python
df_customers = pd.read_csv('customers.csv')
df_customers
```

```python
#1.Wypisz kolumny;
df_customers.columns.tolist()
```

```python
#2.Porównaj rozmiar ramki z rozmiarem ramki w zadaniu 1;
print('Rozmiary ramki z zamówieniami (wiersze, kolumny) to: ', df_orders.shape, 'a rozmiar ramki z klientami to: ', df_customers.shape) #ramka z zamowieniami jest znacznie wieksza, co jest prawdopodobne, poniewaz klient moze robic duzo razy zakupy w sklepie
```

```python
#3.Zmień nazwę kolumny customerID na customer_id upewniając się, że dane zmiana zostanie zapisana w pamięci dzięki jednemu z argumentów metody do zmiany nazwy kolumny;
df_customers.rename(columns = {'customerID': 'customer_id'}, inplace =True)
df_customers
```

```python
#4.Połącz ramkę orders z ramką customers po kolumnie customer_id. Z której metody skorzystasz? Zostaw odpowiedź w komentarzu do zadania w kodzie (np. blok komentarzem ‘’’ ‘’’)
df_order_cust_details = pd.merge(df_orders, df_customers, how='left', on='customer_id')

'''
Wybrałam metodę merge (czyli w języku relacyjnych baz danych join), ponieważ chcę poszerzyć informację o zamówieniach informacjami na temat klienta. Wybrałam typ złączenia left join,
aby nie tracić cennej informacji o zamówieniach dla których nie ma klienta w ramce df_customers. W zależności od potrzebnego formatu danych przydatne może okazać się złączenie typu inner,
jeżeli chcielibyśmy otrzymać informację o zamówieniach złożonych przez tych klientów, którzy znajdują się w ramce df_customers. Wybrane złączenie left join pozwala wyłapać pewny istotny 
problem z danymi w ramce df_customers (liczba wierszy nowej ramki po złaczeniu jest większa niż było w df_orders) - customer_id w tej ramce nie jest unikalny, dla kilkudziesięciu przypadków
mamy różne dane, przykładowo: 
df_customers.loc[df_customers['customer_id'] == '000759']
'''
df_order_cust_details
```
